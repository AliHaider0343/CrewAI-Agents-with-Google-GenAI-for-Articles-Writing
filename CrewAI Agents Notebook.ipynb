# Agents Configuration


```python
from crewai import Agent
from langchain.llms import OpenAI
import os
from langchain_google_genai import ChatGoogleGenerativeAI

os.environ['OPENAI_API_KEY']=""
# Set gemini pro as llm
llm = ChatGoogleGenerativeAI(model="gemini-pro",
                             verbose = True,
                             temperature = 0.5,
                             google_api_key="")

class ArticalsAgents():

  def artical_writer(self):
    return Agent(
        role='Content/Artical Writer',
        goal='Rewrite the Provided Artical English Make sure the Quality is Best.',
        backstory=
        'An expert in analyzing, copywriting, rewritting articals that gets the most attension because of the articals quality.',
        verbose=True,llm=llm)

  def content_optimzer(self):
    return Agent(
        role='SEO and Content Optimzation Expert',
        goal='Optimize artical Written by Content/Artical Writer and extract the keywords.',
        backstory="""A knowledegeable SEO expert that has a vast experince in content optimization and keywords extraction does haev the knowldege about the Technical and non Tecnhical ASpects of Content Optimization.""",
        verbose=True,llm=llm)
```

# Tasks


```python
from crewai import Task
from textwrap import dedent
from datetime import date


class ArticalTasks():

  def RewriteArtical(self, agent, raw_artical):
    return Task(description=dedent(f"""
    Analyze and Rewrite the Given Artical in English and make sure the qulaity of the rewritten artical is top notch.
    if there exist any part of artical other then english lanuage then trasnalte it in to english.
    understand the provided Content and Optimze the Given Artical Add the <strong></strong> or <b></b> tags to make the Important Words Highligted.
      and Must Opitimze the Meta Tags as well as the Keywords from the Artical.
      .

        {self.__tip_section()}
        Raw Artical: {raw_artical}

      """),
                expected_output="""Your Final Answer must be Detailed and Quality Optimzied  and formatted Artical that have hihgliged the important words using the HTML Tags also verify and optimze the keywords and meta Tags and Must follow the below Pattren.
    1. <h1> Article title </h1>
    2. <p> Meta tags </p>
    3. <p> Article Content </p>
    4. <h5> Optimized Keywords </h5>
    5. <p> Summary of Artical </p>

    Donot Add any thing from your own.
      """
                ,agent=agent)

  def optimze_content(self, agent,raw_artical):
    return Task(description=dedent(f"""
      As an Content Optimzation and SEO Expert your Task it to OPtimzie the given Rewritten artical from the Content/Artical Writer.
      Analyze and understand the provided Content and Optimze the Given Artical Add the <strong></strong> or <b></b> tags to make the Important Words Highligted.
      and Must verify and Opitimze the Meta Tags as well as the Keywords provided in the rewritten artical.

        Raw Artical: {raw_artical}

        {self.__tip_section()}

      """),    expected_output="""Your Final Answer must be Detailed and Quality Optimzied  and formatted Artical that have hihgliged the important words using the HTML Tags also verify and optimze the keywords and meta Tags and Must follow the below Pattren.
    1. <h1> Article title </h1>
    2. <p> Meta tags </p>
    3. <p> Article Content </p>
    4. <h5> Optimized Keywords </h5>
    5. <p> Summary of Artical </p>

      Donot Add any thing from your own.

      """
,
                agent=agent)


  def __tip_section(self):
    return "If you do your BEST WORK, I'll tip you $100!"
```

# Execution

```python
from crewai import Crew
from textwrap import dedent

class ArticalsCrew:
  def __init__(self, raw_artical):
    self.raw_artical = raw_artical

  def run(self):
    agents = ArticalsAgents()
    tasks = ArticalTasks()

    artical_writer_agent = agents.artical_writer()
    content_optimzer_agent = agents.content_optimzer()

    RewriteArtical_task = tasks.RewriteArtical(
      artical_writer_agent,
      self.raw_artical
    )

    optimze_content_task = tasks.optimze_content(
      content_optimzer_agent
     )

    crew = Crew(
      agents=[
        artical_writer_agent,content_optimzer_agent
      ],
      tasks=[RewriteArtical_task,optimze_content_task],
      verbose=True
    )

    result = crew.kickoff()
    return result

```

```python
raw_artical = input(
    dedent("""
       what is the Artical you want to Rewrite?
    """))

articalscrew = ArticalsCrew(raw_artical)
result = articalscrew.run()
print("\n\n########################")
print(result)
```

```python
import pandas as pd
import numpy as np
import openai
import os
from langchain.output_parsers import StructuredOutputParser, ResponseSchema
from langchain.prompts import PromptTemplate, ChatPromptTemplate, HumanMessagePromptTemplate
from langchain.llms import OpenAI
from langchain.chat_models import ChatOpenAI
from langchain_google_genai import ChatGoogleGenerativeAI
import google.generativeai as genai
safety_settings_NONE = [
        {"category": "HARM_CATEGORY_HARASSMENT", "threshold": "BLOCK_NONE"},
        {"category": "HARM_CATEGORY_HATE_SPEECH", "threshold": "BLOCK_NONE"},
        {"category": "HARM_CATEGORY_SEXUALLY_EXPLICIT", "threshold": "BLOCK_NONE"},
        {"category": "HARM_CATEGORY_DANGEROUS_CONTENT", "threshold": "BLOCK_NONE"}
    ]

os.environ['GOOGLE_API_KEY'] = 'AIzaSyAJjb0Koe8IdFWQB8jwaVTrwelav20wkMY'
llm = ChatGoogleGenerativeAI(model="gemini-pro")
llm.client = genai.GenerativeModel(model_name='gemini-pro', safety_settings=safety_settings_NONE)

class PromptTemplate:
    def __init__(self, input_variables, template):
        self.input_variables = input_variables
        self.template = template

    def format(self, **kwargs):
        return self.template.format(**kwargs)

response_schemas_for_plg_Analysis = [
    ResponseSchema(
    name="Rewritten Quality Artical",
      description="""
      Rewrite the following article as news article and expand Artical text, bold important words using(<b></b>) and format the article donot compromise the Quality of Artical use headings as well.
      """),
      ResponseSchema(
        name="Summary",
                   description="understandable and quality Summary of Whole Artical."),
      ResponseSchema(
        name="meta Tags",
                   description="Comma Splitted Meta Tags that optimize and clearly dipicts the artical." ),
      ResponseSchema(
        name="keywords",
                   description="Comma Splitted Optimzed key words that could be used for SEO."),
    ResponseSchema(
        name="Important Points",
                   description="write the Important Points from the given Artical in the bUllets for the Video Script."),
  ]

output_parser_for_plg_Conversation = StructuredOutputParser.from_response_schemas(response_schemas_for_plg_Analysis)

format_instructions_for_plg_Analysis = output_parser_for_plg_Conversation.get_format_instructions()

prompt_for_plg= PromptTemplate(
    template="""
    ###Role:
      You are an Artical Rewritter who rewrites the Artical as New Artical in more detial and with high Quality.

    ###Task:
      Your Task is to provide high Quality and expanded Rewritten Artical from the Given Artical.

    ###Responce Instructions:
      1. Must Transalate the Artical parts to english if they are in any other Language.
      2. Must Add the <strong></strong> and <b></b> tag for Important Words and Phareses.
      3. Must user headings and rewrite the artical as News Artical.
      4. Include the Artical Title as well.
      5. Donot use complex words or phrases.
      6. Make it simple to understand.
      7. write everything in appropriate Format as described donot merge variables.

    {format_instructions}

    Given Artical:

    <Raw Artical>{text}</Raw Artical>

    """,
    input_variables=["format_instructions", "text"])

def Get_Inference(text_input):
    formatted_prompt = prompt_for_plg.format(format_instructions=format_instructions_for_plg_Analysis, text=text_input)
    result = (llm.invoke(formatted_prompt))
    return result

```

```python
def process_conversations(text):
    results = Get_Inference(text)
    pairs = results.strip('{}').split(',')
    try:
        pairs = [pair.split(':')[1].replace('"', '').strip() for pair in pairs]
        data = {

                "Rewritten Quality Artical":pairs[0],
                "Summary":pairs[1],
                "meta Tags":pairs[2],
                "keywords":pairs[3],
                "Important Points":pairs[4]
        }
    except Exception as e:
        print(e)
        data = {
             "Rewritten Quality Artical":"Inference Processing Problem.",
              "Summary":"Inference Processing Problem.",
              "meta Tags":"Inference Processing Problem.",
              "keywords":"Inference Processing Problem.",
              "Important Points":"Inference Processing Problem."

        }
    return data
```
